Use method called dummy variables which is helpful in Linear regression

#String Data Categorization to Dummy Variables

```python
#Can assign values such as index
ocean_proximity_dummies = pd.get_dummies(data['ocean_proximity'], prefix = "ocean_proximity").astype(int)
data = pd.concat([data.drop("ocean_proximity", axis=1), ocean_proximity_dummies], axis =1)
ocean_proximity_dummies


```

```python
data = data.drop('ocean_proximity_ISLAND', axis=1)
data.columns
```

```python
#Splitting Data into Train and Test
features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'population', 'households', 'median_income',
       'ocean_proximity_<1H OCEAN', 'ocean_proximity_INLAND',
       'ocean_proximity_NEAR BAY', 'ocean_proximity_NEAR OCEAN']
target = ['median_house_value']
X = data[features]
y = data[target]

#split data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 777) #test is 20% of total data
#training is 80%
#The random state is a random seed
#check size of the splits

print(f'Training set size: {X_train.shape[0]} samples')
print(f'Test set size: {X_test.shape[0]} samples')




```

```python
#Add a constant. Study linear regression to understand
X_train_const = sm.add_constant(X_train)
X_train_const
```

```python
model_fitted = sm.OLS(y_train, X_train_const).fit() #fits the model to the data
#Sets a bias?
```

```python
#print summary
print(model_fitted.summary())
#p values: probabilty to assume if the null hypothesis is true
#t values: compare sample with population data
#tells that the value is independent
#if two houses with different median age. 50 vs 51, the older house would clost $835 more.
#closer to inland ocean proximity costs 2k more
```

#Predicting and Testing


```python
#
X_test_const = sm.add_constant(X_test)

#Make those predictions
test_predictions = model_fitted.predict(X_test_const) #predicts y and x?
test_predictions
```

#Checking OLS Assumptions

## Assumption 1: Linearity?

```python
# Scatter plot for observed vs predicted values on test data
plt.scatter(y_test, test_predictions, color = "forestgreen")
plt.xlabel('Observed Values')
plt.ylabel('Predicted Values')
plt.title('Observed vs Predicted Values on Test Data')
plt.plot(y_test, y_test, color='darkred') # line for perfect prediction (true values)
plt.show()
#shows line of best fit
```

##Assumption 2: Random Sample?

```python
# Calculate the mean of the residuals
mean_residuals = np.mean(model_fitted.resid) #mean of residual. Just prediction - observed. Cannot do a true prediction
#can't have true prediction so using a measure. it's not a observed mean error
#if this is a random sample, the residual should be spread across the zero line
print(f"The mean of the residuals is {np.round(mean_residuals,2)}") #rounds to two decimal places
```

```python
# Plotting the residuals
plt.scatter(model_fitted.fittedvalues, model_fitted.resid, color = "forestgreen")
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()
```

##Assumption 3: Exogeneity (There is no internal correlation)(We have independent values)

```python
# Calculate the residuals
residuals = model_fitted.resid

# Check for correlation between residuals and each predictor
for column in X_train.columns:
  corr_coefficient = np.corrcoef(X_train[column], residuals)[0, 1] #correlation between 0,1
  print(f'Correlation between residuals and {column}: {np.round(corr_coefficient,2)}')
```
